## The goal of this notebook is to visualize the split and statistics of the dataset split

```python
import os
from pathlib import Path
from collections import defaultdict

def count_images(base_dir):
    count_dict = defaultdict(lambda: defaultdict(int))
    for split_dir in base_dir.iterdir():
        if split_dir.is_dir():
            for class_dir in split_dir.iterdir():
                if class_dir.is_dir():
                    num_images = len([f for f in class_dir.iterdir() if f.is_file()])
                    count_dict[split_dir.name][class_dir.name] = num_images
    return count_dict

# Set the path to your dataset
base_dir = Path('../../data/images/2vsall/')

# Get and print the counts
image_counts = count_images(base_dir)
for split, classes in image_counts.items():
    print(f"Split: {split}")
    for class_name, count in classes.items():
        print(f"  Class {class_name}: {count} images")
```

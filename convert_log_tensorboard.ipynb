```python
import json
import os
import torch
from torch.utils.tensorboard import SummaryWriter

# 🔹 Define the log file path
log_file_path = "./output_dir/pretrain/railsemPretrainednomask/log.txt"  # Update with your actual file path

# 🔹 Define TensorBoard writer
tensorboard_log_dir = "./runs/tensorboard_training_logs"  # TensorBoard log directory
writer = SummaryWriter(log_dir=tensorboard_log_dir)

# 🔹 Read and process the log file
with open(log_file_path, "r") as f:
    for line in f:
        data = json.loads(line.strip())  # Parse each JSON line
        
        epoch = data["epoch"]
        train_loss = data["train_loss"]
        train_lr = data["train_lr"]

        # 🔹 Log to TensorBoard
        writer.add_scalar("Loss/train", train_loss, epoch)
        writer.add_scalar("Learning Rate", train_lr, epoch)

# 🔹 Close the writer
writer.close()
```

## RESTRUCTURATION OF DATASET

The goal of this notebook is to restructurate railsem19 dataset regarding a risk classification task.
The formalism follow the cat_dog dataset with the following hierarchy

|TRAIN\
|VAL \
|TEST\
 |\
 |__> |CAT\
 &emsp;&emsp;|DOG



```python
import os
import shutil
import pandas as pd
from pathlib import Path
import random

# Paths
source_folder = '../../data/images/rs19_224/'
csv_path = '../../data/annotation/data_median.csv'
base_folder = '../../data/images/2vsall/'  # This is where the new dataset structure will be created

# Read the CSV file
df = pd.read_csv(csv_path)
```

```python
# Create directories
for split in ['train', 'val', 'test']:
    for category in ['safe', 'risky']:
        os.makedirs(os.path.join(base_folder, split, category), exist_ok=True)
```

```python
# Function to determine risk category
def determine_risk_category(risk_level):
    if risk_level <= 2:
        return 'safe'
    else:
        return 'risky'

# Iterate through the DataFrame
for index, row in df.iterrows():
    file_name = row['Name']  # Assuming the name of the image file is in the 'image_name' column
    risk_level = row['Users']  # Assuming the risk level is in the 'risk_level' column
    dataset_split = row['Fold_0'].str.lower()  # Assuming the dataset split is indicated as 'train', 'val', or 'test'
    
    # Determine the category
    category = determine_risk_category(risk_level)
    
    # Construct source and destination paths
    src_path = os.path.join(source_folder, file_name)
    dest_path = os.path.join(base_folder, dataset_split, category, file_name)
    
    # Copy the file
    shutil.copy2(src_path, dest_path)

print("Dataset restructuring complete!")
```

## HERE CREATE A COPY OF A DATASET REGARDING A PATH BY BALANCING THE CLASSES IN TRAIN TEST VAL 
//
This will result in a smaller dataset

Define here the path of the dataset you want to balance and the path of the resulting balanced dataset

```python
import os
import shutil
import random
from pathlib import Path

source_dir = Path('../../data/images/2vsall')
target_dir = Path('../../data/images/2vsall_bl_2')
```

```python
def balance_and_copy_images(source, target):
    splits = ['train', 'val', 'test']  # Define your dataset splits
    classes = ['risky', 'safe']  # Define your classes

    for split in splits:
        split_source_dir = source / split
        split_target_dir = target / split
        split_target_dir.mkdir(parents=True, exist_ok=True)  # Create train/val/test in target

        # Collect images and determine the minimum count for the current split
        images_dict = {}
        for cls in classes:
            class_path = split_source_dir / cls
            images = list(class_path.glob('*.*'))  # Adjust to your image file extensions
            images_dict[cls] = images

        min_images = min(len(images) for images in images_dict.values())

        # Copy images to balance the classes
        for cls in classes:
            target_class_dir = split_target_dir / cls
            target_class_dir.mkdir(parents=True, exist_ok=True)

            images = images_dict[cls]
            random.shuffle(images)  # Shuffle to randomize which images are selected

            # Copy the balanced number of images
            for image in images[:min_images]:
                shutil.copy(image, target_class_dir)

# Run the function for your dataset
balance_and_copy_images(source_dir, target_dir)
```

```python
balance_and_copy_images(source_dir, target_dir)
```
